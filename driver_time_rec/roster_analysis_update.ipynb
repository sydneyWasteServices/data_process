```python
PATH_ROSTER = 'D:\\Run Analysis\\BLOB_STORAGE\\Roster\\weekly_roster_processed\\14th_2021_roster.xlsx'
PATH_SALARY = 'D:\\Run Analysis\\BLOB_STORAGE\\Driver_weekly_timesheet\\processed_timesheet\\14th_2021_timesheet.csv'

df_sal = pd.read_csv(PATH_SALARY)
df_ros = pd.read_excel(PATH_ROSTER)
df_ros.groupby('Primary_employeeID')['Primary_driver_Name'].count()

# how='left'
df_sal_ros = pd.merge(df_ros, df_sal , left_on='Primary_employeeID',right_on='EmployeeID_y')

df_sal_ros[[
    'Date',
    'Run_type',
    'Primary_employeeID',
    'Secondary_employeeID',
    'Primary_driver_Name',
    'Secondary_Driver_Name',
    'Primary_route',
    'Satellite_Route_1',
    'Satellite_Route_2',
    'Primary_truck',
    'Alternative_Truck',
    'Subcontracted_From/Special_Client',
    'Total_hour'
]]

occurence_per_runs_ = (df_sal_ros
 .groupby(['Primary_employeeID', 'Primary_route','Total_hour'])['Run_type']
 .count()
 .reset_index())
# 56,887.79 
total_runs = df_sal_ros.groupby('Primary_employeeID')['Run_type'].count().reset_index()

df_1 = pd.merge(occurence_per_runs_, total_runs, on='Primary_employeeID', how='left')

df_ab= df_1.drop_duplicates(subset='Primary_employeeID',keep='first')
t_hours = df_ab.Total_hour.sum()

avg_rate = 56887.79 / t_hours

# df_1.transform(lambda df : df['Total_hour'] * (df['Run_type_x']/df['Run_type_y']) )
# df_1.apply(lambda df : df['Total_hour'] * (df['Run_type_x']/df['Run_type_y']))


df_1['hours_per_run'] = df_1.pipe(lambda df :df['Total_hour'] * (df['Run_type_x']/df['Run_type_y']) )
df_1.pipe(lambda df :df['Total_hour'] * (df['Run_type_x']/df['Run_type_y']) )

df_1['driver_salary_per_run'] = df_1.pipe(lambda df :df['hours_per_run'] * avg_rate  )
df_1.sort_values('Primary_route')

df_1.groupby('Primary_route')['driver_salary_per_run'].sum().reset_index()

df_1['hours_per_run'] = df_1['Run_type_x'] / df_1['Run_type_y'] * df_1['Total_hour']

df_1.hours_per_run.sum()

# CAMP

# HOOK => HOOK1

# UOS

# UOS CAMDEN

# UOS CB

```
