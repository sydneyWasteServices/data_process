```python
PATH_SUEZ = 'D:\\Run Analysis\\BLOB_STORAGE\\tipping_weekly\\og_tipping_suez\\weekly\\18th_2021.csv' 
PATH_ROS = 'D:\\Run Analysis\\BLOB_STORAGE\\Roster\\weekly_roster_processed\\18th_2021.xlsx' 
PATH_WE = 'D:\\Run Analysis\\BLOB_STORAGE\\tipping_weekly\\waste_edge_tipping\\18th_2021.csv'
df_suez = pd.read_csv(PATH_SUEZ)
df_ros = pd.read_excel(PATH_ROS)
df_we = pd.read_csv(PATH_WE)

GENERAL_WASTE = ['HOOK1', 'BR1', 'BR2', 'BR3', 'FL2', 'FLG', 'RL1', 'RL2', 'RL4', 'RL7', 'RL9', 'RLD', 'RLE', 'RLH', 'RLI', 'RLJ', 'RLK', 'SWG']
NOTGW_RUN = ['APR', 'FLP', 'HYG', 'RED', 'RL5', 'RL6', 'RL8', 'RLP', 'RLR', 'SWP', 'HOOKCB','CBK', 'RLC', 'RLG', 'DOY','NEPCB', 'UOSCB','UOSCO','CMDCB','CUMCB']
UOS = ['UOSGW', 'CMDGW', 'CUMGW', 'NEPGW']

# Date 
df_suez.Date = pd.to_datetime(df_suez.Date, format='%d/%m/%Y')
df_we[['Route No','weekday']]= df_we['Route No'].str.split('-', 1, expand=True)
df_we['Route Date'] = pd.to_datetime(df_we['Route Date'], format='%d-%b-%Y', errors='coerce')
df_we['Disposal Date'] = pd.to_datetime(df_we['Disposal Date'], format='%d-%b-%Y', errors='coerce')
# =====================================

# Filter No Gross and Tare Weight All & Non_GW_run
df_we = df_we.loc[(df_we['Gross Weight'] > 0) & df_we['Tare Weight']]
df_gw_runs = df_we[~df_we['Route No'].isin(NOTGW_RUN)]
df_gw_runs = df_gw_runs.drop(columns=['Route Date', 'Disposal Date','Total Charge','Cost Rate','Charge Rate','UOM','Branch', 'Note','Booking No','Customer Details'])    

# Clean Dockets 
# df_gw_runs['Docket No'].
df_gw_runs['Docket No'] = df_gw_runs['Docket No'].str.replace('.0','',regex=False).str.replace(' ','')   

# Inspect Any Duplicate in Waste Edge Tipping
df_gw_runs[df_gw_runs.duplicated(subset=['Docket No'])]
# Inspect Any Duplicate in Suez Tipping
df_suez[df_suez.duplicated(subset=['Docket'])]
```
