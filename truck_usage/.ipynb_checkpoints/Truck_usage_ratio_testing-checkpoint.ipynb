```python
import pandas as pd
import numpy as np
import glob
```

```python
PATH_OLDROS = 'D:\\Run Analysis\\BLOB_STORAGE\\Roster\\Rearranged_past_roster\\past_ros.csv'
PATH_CURROS = 'D:\\Run Analysis\\BLOB_STORAGE\\Roster\\weekly_roster_processed'
df_old_ros = pd.read_csv(PATH_OLDROS)
df_old_ros.Date = df_old_ros.Date.str.split(' ', 1 ,expand=True)[0]
df_old_ros.Date = pd.to_datetime(df_old_ros.Date, format='%d/%m/%Y')

df_old_ros.Primary_truck = df_old_ros.Primary_truck.str.replace(' ', '')
df_old_ros = df_old_ros.loc[df_old_ros.Date <= '20210406' ]


df = pd.concat(map(pd.read_excel, glob.glob(f'{PATH_CURROS}//*.xlsx')))
df.Primary_truck = df.Primary_truck.str.replace(' ', '')

df_combine = pd.concat([df, df_old_ros])
df_combine = df_combine.sort_values('Date')
```

```python
# COMPLETE_PATH = 'D:\\Run Analysis\\BLOB_STORAGE\\truckUsage\\17th_2021.csv'
# .reset_index().to_csv(COMPLETE_PATH, index=False)
occurence = df_combine.groupby(['Primary_truck','Primary_route'])['Run_type'].count().reset_index()
Total = df_combine.groupby('Primary_truck')['Run_type'].count().reset_index()
```

```python
df_merge = pd.merge(occurence, Total, how='left', on='Primary_truck')
df_merge['ratio'] = df_merge['Run_type_x'] / df_merge['Run_type_y']
mv_exp_ratio_table = df_merge[['Primary_truck','Primary_route','ratio']].sort_values('Primary_route')
```

```python
PATH_MVEXP = 'D:\\Run Analysis\\BLOB_STORAGE\\expenses_truck\\all_other_mv_expense\\to_be_process\\17th_2021.xlsx'
PATH_ROS = 'D:\\Run Analysis\\BLOB_STORAGE\\Roster\\weekly_roster_processed\\17th_2021.xlsx'
PATH_OLDROS = 'D:\\Run Analysis\\BLOB_STORAGE\\Roster\\Rearranged_past_roster\\All_Past_Sorted_ROSTER.csv'

df_mvexp = pd.read_excel(PATH_MVEXP)
df_ros = pd.read_excel(PATH_ROS)

df_mvexp.iloc[:,3] = pd.to_datetime(df_mvexp.iloc[:,3], format="%d/%m/%Y")
# df_mvexp.groupby('Job No.')['Debit'].sum().reset_index()
mv_exp_table = df_mvexp.groupby(['catID','Job No.'])['Debit'].sum().reset_index()
# mv_exp_table.Debit.sum()
```

```python
df_merge1 = pd.merge(mv_exp_table, mv_exp_ratio_table, how='left', left_on='Job No.',right_on='Primary_truck')
df_merge1['amount_portion'] = df_merge1['Debit'] * df_merge1['ratio']
```

```python
ratio_table = df_merge1['Job No.'].unique()
exp_table = mv_exp_table['Job No.'].unique()
```
